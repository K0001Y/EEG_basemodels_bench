# Six-category radar plot for linear_full

Set `SPLIT` to `"cross"` or `"within"`, then run the cell. The plot style matches `linear_downsample_cross_radar.ipynb`.

```python
from pathlib import Path
import runpy
import sys
from IPython.display import Image, display

SPLIT = 'cross'  # 'cross' or 'within'

analysis_dir = Path('/benchmark-eeg/5.0_version/analysis/six_category_radar')
sys.argv = [
    'linear_full_radar.py',
    '--split', SPLIT,
]
runpy.run_path(str(analysis_dir / 'linear_full_radar.py'), run_name='__main__')
display(Image(filename=str(analysis_dir / f'linear_full_{SPLIT}_six_category_radar.png')))
```
